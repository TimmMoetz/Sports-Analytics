#### B) Vektoren Funktion

```python
def vector_ABC(df): 
    r_foot=[]
    l_foot=[]
    #Fuß
    d27_29=[]
    d27_31=[]
    d29_31=[]
    d28_30=[]
    d28_32=[]
    d30_32=[]
    #Hüfte, Knie
    d23_25=[]
    d25_27=[]
    d24_26=[]
    d26_28=[]

    for i in df.index:
        #rechter Fuß
        p27 = np.array([df.iloc[i,12],df.iloc[i,13],df.iloc[i,14]])
        p29 = np.array([df.iloc[i,18],df.iloc[i,19],df.iloc[i,20]])
        p31 = np.array([df.iloc[i,24],df.iloc[i,25],df.iloc[i,26]])
        #rechte Hüfte, Knie
        p23 = np.array([df.iloc[0,0],df.iloc[0,1],df.iloc[0,2]])
        p25 = np.array([df.iloc[0,6],df.iloc[0,7],df.iloc[0,8]])
        #linker Fuß
        p28 = np.array([df.iloc[i,15],df.iloc[i,16],df.iloc[i,17]])
        p30 = np.array([df.iloc[i,21],df.iloc[i,22],df.iloc[i,23]])
        p32 = np.array([df.iloc[i,27],df.iloc[i,28],df.iloc[i,29]])
        #linke Hüfte, Knie
        p24 = np.array([df.iloc[0,3],df.iloc[0,4],df.iloc[0,5]])
        p26 = np.array([df.iloc[0,9],df.iloc[0,10],df.iloc[0,11]])
        #Verbindungsvektoren Fuß
        ABr = p29 - p27
        ACr = p31 - p27
        ABl = p30 - p28
        ACl = p32 - p28
        #Kreuzprodukt
        kpr = np.cross(ABr,ACr)
        kpl = np.cross(ABl,ACl)
        #Länge des Vektors
        lr = np.linalg.norm(kpr)
        ll = np.linalg.norm(kpl)
        #Ebene
        ABCr = lr/2
        ABCl = ll/2

        r_foot.append(ABCr)
        l_foot.append(ABCl)

        #Abstände
        d23_25.append(np.linalg.norm(p23-p25))
        d25_27.append(np.linalg.norm(p25-p27))
        d27_29.append(np.linalg.norm(p27-p29))
        d27_31.append(np.linalg.norm(p27-p31))
        d29_31.append(np.linalg.norm(p29-p31))

        d24_26.append(np.linalg.norm(p24-p26))
        d26_28.append(np.linalg.norm(p26-p25))
        d28_30.append(np.linalg.norm(p28-p30))
        d28_32.append(np.linalg.norm(p28-p32))
        d30_32.append(np.linalg.norm(p30-p32))



    df["r_foot"]=r_foot
    df["l_foot"]=l_foot
    
    df["d23_25"]=d23_25
    df["d25_27"]=d25_27
    df["d24_26"]=d24_26
    df["d26_28"]=d26_28

    df["d27_29"]=d27_29
    df["d27_31"]=d27_31
    df["d29_31"]=d29_31
    df["d28_30"]=d28_30
    df["d28_32"]=d28_32
    df["d30_32"]=d30_32
    
```

#### D) IQR Ausreißer
> Wurde nach dem Coaching nicht mehr verwendet

```python
def IQR_frame_error(df):
    error_frames=0
    upper_error=[]
    lower_error=[]
    for i in df.iloc[:,-12:-6]:
        Q1 = np.percentile(df[i], 25, interpolation = 'midpoint')
        Q3 = np.percentile(df[i], 75, interpolation = 'midpoint')
        IQR = Q3 - Q1
        upper = df[i] >= (Q3+1.5*IQR)
        lower = df[i] <= (Q1-1.5*IQR)
        if not np.any(upper_error):
            upper_error=np.where(upper)
        else:
            upper_error=np.unique(np.concatenate((upper_error,np.where(upper)),None))
        if not np.any(lower_error):
            lower_error=np.where(lower)
        else:
            lower_error=np.unique(np.concatenate((lower_error,np.where(lower)),None))
    
    print(upper_error,lower_error)
    error_frames=len(upper_error)+len(lower_error)
    return error_frames
```

#### E) STD3 Ausreißer

```python
def STD3_frame_error(df):
    error_frames=0
    upper_error=[]
    lower_error=[]
    for i in df.iloc[:,-12:-6]:
        upper_inl = df[i].mean() + df[i].std()*3
        lower_inl = df[i].mean() - df[i].std()*3
        upper = df[i] >= upper_inl
        lower = df[i] <= lower_inl
        if not np.any(upper_error):
            upper_error=np.where(upper)
        else:
            upper_error=np.unique(np.concatenate((upper_error,np.where(upper)),None))
        if not np.any(lower_error):
            lower_error=np.where(lower)
        else:
            lower_error=np.unique(np.concatenate((lower_error,np.where(lower)),None))
        
    if np.any(upper_error):
        error_frames=len(upper_error)
    if np.any(lower_error):
        error_frames=error_frames+len(lower_error)
    
    return error_frames
```
