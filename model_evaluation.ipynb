```python
p27 = [df.iloc[0,0],df.iloc[0,1],df.iloc[0,2]]
p29 = [df.iloc[0,6],df.iloc[0,7],df.iloc[0,8]]
p31 = [df.iloc[0,12],df.iloc[0,13],df.iloc[0,14]]
print(p27)
print(p29)
print(p31)
```

```python
vA = np.array(p27)
vB = np.array(p29)
vC = np.array(p31)

#Verbindungsvektoren
AB = vB - vA
AC = vC - vA

#Kreuzprodukt
kp = np.cross(AB,AC)

#LÃ¤nge des Vektors
lae = np.linalg.norm(kp)

#durch 2
ABC = lae/2

ABC
```

#### B) Vektoren Funktion

```python
def vector_ABC(df): 
    r_foot=[]
    l_foot=[]
    d27_29=[]
    d27_31=[]
    d29_31=[]
    d28_30=[]
    d28_32=[]
    d30_32=[]

    for i in df.index:
        #rechter FuÃŸ
        p27 = np.array([df.iloc[i,0],df.iloc[i,1],df.iloc[i,2]])
        p29 = np.array([df.iloc[i,6],df.iloc[i,7],df.iloc[i,8]])
        p31 = np.array([df.iloc[i,12],df.iloc[i,13],df.iloc[i,14]])
        #linker FuÃŸ
        p28 = np.array([df.iloc[i,3],df.iloc[i,4],df.iloc[i,5]])
        p30 = np.array([df.iloc[i,9],df.iloc[i,10],df.iloc[i,11]])
        p32 = np.array([df.iloc[i,15],df.iloc[i,16],df.iloc[i,17]])
        #Verbindungsvektoren
        ABr = p29 - p27
        ACr = p31 - p27
        ABl = p30 - p28
        ACl = p32 - p28
        #Kreuzprodukt
        kpr = np.cross(ABr,ACr)
        kpl = np.cross(ABl,ACl)
        #LÃ¤nge des Vektors
        lr = np.linalg.norm(kpr)
        ll = np.linalg.norm(kpl)
        #Ebene
        ABCr = lr/2
        ABCl = ll/2

        r_foot.append(ABCr)
        l_foot.append(ABCl)

        #AbstÃ¤nde
        d27_29.append(np.linalg.norm(p27-p29))
        d27_31.append(np.linalg.norm(p27-p31))
        d29_31.append(np.linalg.norm(p29-p31))

        d28_30.append(np.linalg.norm(p28-p30))
        d28_32.append(np.linalg.norm(p28-p32))
        d30_32.append(np.linalg.norm(p30-p32))



    df["r_foot"]=r_foot
    df["l_foot"]=l_foot
    df["d27_29"]=d27_29
    df["d27_31"]=d27_31
    df["d29_31"]=d29_31
    df["d28_30"]=d28_30
    df["d28_32"]=d28_32
    df["d30_32"]=d30_32
```

#### D) IQR AusreiÃŸer

```python
def IQR_frame_error(df):
    error_frames=0
    upper_error=[]
    lower_error=[]
    for i in df.iloc[:,-8:-6]:
        Q1 = np.percentile(df[i], 25, interpolation = 'midpoint')
        Q3 = np.percentile(df[i], 75, interpolation = 'midpoint')
        IQR = Q3 - Q1
        upper = df[i] >= (Q3+1.5*IQR)
        lower = df[i] <= (Q1-1.5*IQR)
        if not upper_error:
            upper_error=np.where(upper)
        else:
            upper_error=np.unique(np.concatenate((upper_error,np.where(upper)),None))
        if not lower_error:
            lower_error=np.where(lower)
        else:
            lower_error=np.unique(np.concatenate((lower_error,np.where(lower)),None))

    error_frames=len(upper_error)+len(lower_error)
    return error_frames
```

```python
IQR_frame_error(df)
```

#### E) STD3 AusreiÃŸer

```python
def STD3_frame_error(df):
    error_frames=0
    upper_error=[]
    lower_error=[]
    for i in df.iloc[:,-8:-6]:
        upper_inl = df[i].mean() + df[i].std()*3
        lower_inl = df[i].mean() - df[i].std()*3
        upper = df[i] >= upper_inl
        lower = df[i] <= lower_inl
        if not upper_error:
            upper_error=np.where(upper)
        else:
            upper_error=np.unique(np.concatenate((upper_error,np.where(upper)),None))
        if not lower_error:
            lower_error=np.where(lower)
        else:
            lower_error=np.unique(np.concatenate((lower_error,np.where(lower)),None))
        
    error_frames=len(upper_error)+len(lower_error)
    return error_frames
```
