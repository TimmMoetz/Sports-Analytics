### All Data


```python
r_foot=[]
l_foot=[]
d27_29=[]
d27_31=[]
d29_31=[]
d28_30=[]
d28_32=[]
d30_32=[]

for i in df.index:
    #rechter Fuß
    p27 = np.array([df.iloc[i,0],df.iloc[i,1],df.iloc[i,2]])
    p29 = np.array([df.iloc[i,6],df.iloc[i,7],df.iloc[i,8]])
    p31 = np.array([df.iloc[i,12],df.iloc[i,13],df.iloc[i,14]])
    #linker Fuß
    p28 = np.array([df.iloc[i,3],df.iloc[i,4],df.iloc[i,5]])
    p30 = np.array([df.iloc[i,9],df.iloc[i,10],df.iloc[i,11]])
    p32 = np.array([df.iloc[i,15],df.iloc[i,16],df.iloc[i,17]])
    #vektoren
    vrA = np.array(p27)
    vrB = np.array(p29)
    vrC = np.array(p31)
    vlA = np.array(p28)
    vlB = np.array(p30)
    vlC = np.array(p32)
    #Verbindungsvektoren
    ABr = p29 - p27
    ACr = p31 - p27
    ABl = p30 - p28
    ACl = p32 - p28
    #Kreuzprodukt
    kpr = np.cross(ABr,ACr)
    kpl = np.cross(ABl,ACl)
    #Länge des Vektors
    lr = np.linalg.norm(kpr)
    ll = np.linalg.norm(kpl)
    #Ebene
    ABCr = lr/2
    ABCl = ll/2

    r_foot.append(ABCr)
    l_foot.append(ABCl)

    #Abstände
    d27_29.append(np.linalg.norm(p27-p29))
    d27_31.append(np.linalg.norm(p27-p31))
    d29_31.append(np.linalg.norm(p29-p31))

    d28_30.append(np.linalg.norm(p28-p30))
    d28_32.append(np.linalg.norm(p28-p32))
    d30_32.append(np.linalg.norm(p30-p32))

    

df["r_foot"]=r_foot
df["l_foot"]=l_foot
df["d27_29"]=d27_29
df["d27_31"]=d27_31
df["d29_31"]=d29_31
df["d28_30"]=d28_30
df["d28_32"]=d28_32
df["d30_32"]=d30_32
```

> Anomalien treten insbesondere beim rechten Fuß auf. Bei genauerer Betrachtung der Abstände der einzelnen Keypoints scheint insbesonder p31 (rechte Fußspitze) zu beginn des Videos Probleme zu machen

```python
Q1 = np.percentile(df["r_foot"], 25, interpolation = 'midpoint')
Q3 = np.percentile(df["r_foot"], 75, interpolation = 'midpoint')
IQR = Q3 - Q1
upper = df["r_foot"] >= (Q3+1.5*IQR)
lower = df["r_foot"] <= (Q1-1.5*IQR)

error_frames = np.size(np.where(lower)) + np.size(np.where(upper))

print("Anazhl Fehlerhafter Frames Aufgrund der Abstände im rechten Fuß:", error_frames)
```
