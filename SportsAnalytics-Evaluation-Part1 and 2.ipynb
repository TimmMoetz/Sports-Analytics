<h3>Detect Outlier<h3>

```python
#z-score technique
outliers = []
def detect_outlier(df):

    threshold=3
    mean = np.mean(df)
    std =np.std(df)

    for i in df:
        z_score = (i-mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers

outlier = detect_outlier(df["24x"])
print(outlier)
```

```python
#STD3-Technique
def STD3_frame_error(df):
    error_frames=0
    upper_error=[]
    lower_error=[]
    for i in df.iloc[:,1:]:
        upper_inl = df[i].mean() + df[i].std()*3
        lower_inl = df[i].mean() - df[i].std()*3
        upper = df[i] >= upper_inl
        lower = df[i] <= lower_inl
        if not np.any(upper_error):
            upper_error=np.where(upper)
        else:
            upper_error=np.unique(np.concatenate((upper_error,np.where(upper)),None))
        if not np.any(lower_error):
            lower_error=np.where(lower)
        else:
            lower_error=np.unique(np.concatenate((lower_error,np.where(lower)),None))
        
    if np.any(upper_error):
        error_frames=len(upper_error)
    if np.any(lower_error):
        error_frames=error_frames+len(lower_error)
    
    return error_frames
```

```python
files_dir=os.listdir("Keypoints")

error_gehen=[]
error_joggen=[]
error_laufen=[]

for file in files_dir:
    with open(f'Keypoints/{file}', 'r') as read:
        df = pd.read_csv(read, sep=',') 
        df
        error=STD3_frame_error(df) 
        if "gehen" in file:
            error_gehen.append(error)
        elif "joggen" in file:
            error_joggen.append(error)
        elif "laufen" in file:
            error_laufen.append(error)  
        else:
            print(file)
            print("Something Wrong!") 
        
print("error-frame gehen: ", sum(error_gehen))
print("error-frame joggen: ", sum(error_joggen))
print("error-frame laufen: ", sum(error_laufen))
```

```python
files_dir=os.listdir("Keypoints")

personen={}

for i in range (1,11):
    personen[f'{i}_']=[]

for file in files_dir:
    with open(f'Keypoints/{file}', 'r') as read:
        df = pd.read_csv(read, sep=',') 
        df
        error=STD3_frame_error(df)
        for i in personen:
            if str(i) in file:
                personen[i].append(error)

print ("Falsche Frames pro Person:\n",personen)   
  
for i in personen:
    personen[i]=sum(personen[i])

print ("Anzahl der falschen frames pro Person\n",personen)  
```
