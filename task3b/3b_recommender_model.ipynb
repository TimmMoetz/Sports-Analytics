```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

```python
def most_common(lst):
    return max(set(lst), key=lst.count)
```

```python
def get_mean_laufstil():
    path = r'task-3b/metrics_laufstil'                     
    all_files = glob.glob(os.path.join(path, "*.csv"))
    df_from_each_file = (pd.read_csv(f) for f in all_files)
    concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)
    return concatenated_df.mean()
```

```python
def get_laufstil(df, filename):
    liste_laufstil_l = []
    liste_laufstil_r = []
    accuracy_l = []
    accuracy_r = []
    mean_r = get_mean_laufstil()["right"]
    mean_l = get_mean_laufstil()["left"]


    for index, row in df.iterrows():
        diff = abs(row["left"] - mean_l)
        accuracy_l.append(diff)
        if row["left"] == mean_l:
            liste_laufstil_l.append("gleich_mean")
        elif row["left"] > mean_l:
            liste_laufstil_l.append("größer_mean")
        else:
            liste_laufstil_l.append("kleiner_mean")
    for index, row in df.iterrows():
        diff = abs(row["right"] - mean_r)
        accuracy_r.append(diff)
        if row["right"] == mean_r:
            liste_laufstil_r.append("gleich_mean")
        elif row["right"] > mean_r:
            liste_laufstil_r.append("größer_mean")
        else:
            liste_laufstil_r.append("kleiner_mean")
                          
                
    data_laufstil = {
        "runner_laufstil": filename[:14],
        "laufstil_r": most_common(liste_laufstil_r),
        "diff_r": most_common(accuracy_r),
        "laufstil_l": most_common(liste_laufstil_l),
        "diff_l": most_common(accuracy_l),
    }
    
    
    return data_laufstil
```

```python
def get_pronation(df, filename):
    liste_pronation_l = []
    liste_pronation_r = []
    accuracy_l = []
    accuracy_r = []
    for index, row in df.iterrows():
        if row["left"] <= 165:
            diff = row["left"] - 165
            accuracy_l.append(diff)
            liste_pronation_l.append("Überpronation")
        elif row["left"] > 165 and row["left"] < 195:
            diff = row["left"] - 180
            accuracy_l.append(diff)
            liste_pronation_l.append("normal_Pronation")
        else:
            diff = row["left"] - 195
            accuracy_l.append(diff)
            liste_pronation_l.append("Unterpronation")
            
    for index, row in df.iterrows():
        if row["right"] <= 165:
            diff = row["right"] - 165
            accuracy_r.append(diff)
            liste_pronation_r.append("Überpronation")
        elif row["right"] > 165 and row["right"] < 195:
            diff = row["right"] - 180
            accuracy_r.append(diff)
            liste_pronation_r.append("normal_Pronation")
        else:
            diff = row["right"] - 195
            accuracy_r.append(diff)
            liste_pronation_r.append("Unterpronation")
        
    
    data_pronation = {
        "runner_pronation": filename[:14],
        "pronation_r": most_common(liste_pronation_r),
        "diff_r": most_common(accuracy_r),
        "pronation_l": most_common(liste_pronation_l),
        "diff_l": most_common(accuracy_l),
    }
    
    
    return data_pronation
    
```

```python
def get_pronation_dir():
    dir_pronation = os.path.abspath("task-3b/metrics_pronation")

    pronation_dir = []

    # loop through all videos with perspective "seite"
    for filename in os.listdir(dir_pronation):
        df = pd.read_csv("task-3b/metrics_pronation/" + filename, index_col=0)
        pronation_dir.append(get_pronation(df, filename))
    pronation_dir = pd.DataFrame(pronation_dir)
    return pronation_dir
```

```python
def get_laufstil_dir():
    dir_laufstil = os.path.abspath("task-3b/metrics_laufstil")
    
    laufstil_dir = []

    # loop through all videos with perspective "seite"
    for filename in os.listdir(dir_laufstil):
        df = pd.read_csv("task-3b/metrics_laufstil/" + filename, index_col=0)
        laufstil_dir.append(get_laufstil(df, filename))
    laufstil_dir = pd.DataFrame(laufstil_dir)
    return laufstil_dir
```

```python
result = pd.concat([get_laufstil_dir(), get_pronation_dir()], axis=1)
result
```
